```python
#Importing al the required Libraries..
import pandas as pd
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
use_cols = [
    'loan_amnt', 'int_rate', 'annual_inc', 'open_acc', 'loan_status',
    'open_il_12m'
]

data = pd.read_csv(
    'loan.csv', usecols=use_cols).sample(
        10000, random_state=44)  # set a seed for reproducibility

data.head()
```

```python
# Visualizing a Contineous Variable
# Here Lone amount are in US Dollars
# Lets Retrive the Unique Amount borrowed By the people.
loan_amount=data.loan_amnt.unique()
print(loan_amount)
```

```python
# Ploting a histogram using Seaborn
sns.set(color_codes=True)
Lp=sns.distplot(loan_amount, kde=False, rug=True,)
plt.title("Loan Amount Plotting")
plt.xlabel("Loan Amount")
plt.ylabel("% of People")
```

```python
Interest_rate=data.int_rate.unique()
print(Interest_rate)
```

```python
# Ploting a histogram using Matplotlib
Lp=data.int_rate.hist(bins=30)
plt.title("Interest Rate Plotting")
plt.xlabel("Interst Rate")
plt.ylabel("Number of People of People")
```

```python
# Plotting the anual Income
Ai = data.annual_inc.hist(bins=100)
Ai.set_xlim(0, 400000)
Ai.set_title("Customer's Annual Income")
Ai.set_xlabel('Annual Income')
Ai.set_ylabel('Number of Customers')
```

```python
# let's create one additional variable called defaulted.
# This variable indicates if the loan has defaulted, which means,
# if the borrower failed to re-pay the loan, and the money
# is deemed lost.

data['defaulted'] = np.where(data.loan_status.isin(['Default']), 1, 0)
data.defaulted.mean()
```

```python
# A binary variable, can take 2 values. For example,
# the variable defaulted that we just created:
# either the loan is defaulted (1) or not (0)

data.defaulted.unique()
```

```python
# let's make a histogram, although histograms for
# binary variables do not make a lot of sense

fig = data.defaulted.hist()
fig.set_xlim(0, 2)
fig.set_title('Defaulted accounts')
fig.set_xlabel('Defaulted')
fig.set_ylabel('Number of Loans')
```
